# Parte 1 – Questão Teórica

A arquitetura apresentada no diagrama é baseada em microsserviços e possui o objetivo de permitir comunicação segura, escalável e monitorada entre clientes, APIs e serviços internos.

### Componentes principais:

**Cloud API**
Responsável por expor os serviços principais da aplicação para clientes externos.

**Gateway (Kong / Reverse Proxy)**
Centraliza o acesso aos microsserviços, realizando roteamento, segurança, autenticação e controle de tráfego.

**Microsserviços**
Cada microsserviço possui uma responsabilidade específica, como processamento de dados, regras de negócio ou comunicação com sistemas externos.

**Storage Service (S3 abstraction)**
Camada intermediária responsável por abstrair a comunicação com o armazenamento S3, aumentando a segurança e desacoplando a aplicação do provedor de storage.

**Banco de Dados**
Armazena informações persistentes da aplicação, como usuários e registros operacionais.

**Prometheus**
Responsável pela coleta de métricas de performance e monitoramento.

**Grafana**
Utilizado para visualização das métricas coletadas pelo Prometheus através de dashboards.

# Parte 3 – Integração com Modelo de Inteligência Artificial

Para demonstrar a integração de IA com a API, foi desenvolvido um modelo simples de Machine Learning utilizando regressão linear com Scikit-Learn.

O objetivo do modelo é prever valores numéricos com base em entradas fornecidas.

Fluxo da implementação:

1. Criação dos dados de treino
2. Treinamento do modelo de regressão linear
3. Teste de predição
4. Salvamento do modelo em arquivo `.pkl` com Joblib
5. Carregamento do modelo na API via endpoint `/predict`


```python
from sklearn.linear_model import LinearRegression
import numpy as np
import joblib
```

## Criação dos dados de treino

Foram definidos dados simples de entrada (X) e saída esperada (y), simulando uma relação linear.


```python
# X = entradas
X = np.array([[1], [2], [3], [4], [5]])

# y = saídas esperadas
y = np.array([2, 4, 6, 8, 10])

print("Dados criados com sucesso")
print("X =", X)
print("y =", y)
```

## Treinamento do modelo

O modelo foi treinado utilizando regressão linear para aprender o padrão entre entradas e saídas.


```python
# cria o modelo de regressão linear
model = LinearRegression()

# treina o modelo com nossos dados
model.fit(X, y)

print("Modelo treinado com sucesso!")
```

## Teste de predição

Após o treinamento, foi realizado um teste de inferência para validar o comportamento do modelo.


```python
prediction = model.predict([[9]])

print("Predição:", prediction)
```

## Persistência do modelo

Após o treinamento, o modelo foi salvo em disco no formato `.pkl`, permitindo que a API carregue o modelo sem necessidade de novo treinamento.


```python
joblib.dump(model, "../app/ml_model.pkl")

print("Modelo salvo com sucesso!")
```
